```python
sal_by_dept = df.groupby('dept')["Salary"].sum()
sal_by_dept
```

```python
# bar plot
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)
```

```python
plt.bar(["HR", "IT", "Finance"], [hr_mean, it_mean, fin_mean], color = ["green", "black", "red"])
plt.show()
```

```python
#Multivariate Analysis: 3 nemeric colums
df.head()
```

```python
df['Exp'] = [1,1.5,2,2.5]*2
df
```

```python
#Bubble plot
plt.scatter(df['Age'],df['Salary'],s = df['Exp']*20,color = "orange",edgecolor="black")
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

```python
#2 numerical and 1 categorical column:
```

```python
plt.scatter(df['Age'],df['Salary'], c =df['dept'].map({"HR":"Yellow","IT":"red","Finance":"Green"}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
color = {"HR": "yellow", "IT": "blue", "Finance": "orange"}
for dept, color in color.items():
  df_dept = df[df["dept"] == dept]
  plt.scatter(df_dept["Age"], df_dept["Salary"], c =color, label= dept)
plt.legend()
```

```python
#Object oriented API
```

```python
fig,axs = plt.subplots(1,3,figsize = (20,5)) #In one fig i want's 3 subplots
#Line plot
axs[0].plot(sort_age["Age"], df["Salary"], color = "red", marker="*",linewidth = "2")
axs[0].grid()
axs[0].set_title("Line Plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")


# histogram:
axs [1].hist(df["Salary"], bins =3, color="Pink")
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")

# Boxplot:
axs [2].boxplot(df["Salary"])
axs [2].set_title("Boxplot")
axs [2].set_xlabel("Salary")

#This the statement to save the ploted graph in our local device
plt.savefig("Muntiple_plots.png")

plt.show()
```
